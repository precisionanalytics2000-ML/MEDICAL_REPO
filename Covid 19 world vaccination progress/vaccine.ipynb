```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import os
```

```python
%pip install --upgrade nbformat
```

```python
df = pd.read_csv('country_vaccinations.csv/country_vaccinations.csv')
```

```python
df
```

```python

df['date'] = pd.to_datetime(df['date'])
df = df[df['country'].isin(['United States', 'India', 'Brazil', 'United Kingdom', 'China'])]  # Top countries
```

```python
df_pivot = df.pivot(index='date', columns='country', values='people_vaccinated_per_hundred')
```

```python
os.makedirs('visuals', exist_ok=True)
os.makedirs('excel_files', exist_ok=True)
```

```python
fig = go.Figure()
for col in df_pivot.columns:
    fig.add_trace(go.Scatter(x=df_pivot.index, y=df_pivot[col], mode='lines', name=col))
fig.update_layout(title='Vaccination Trends (Top Countries)', xaxis_title='Date', yaxis_title='Vaccinated %')
fig.write_html('visuals/vax_trends.html')  # Interactive for GitHub
fig.show()
```

```python
latest = df.groupby('country')['people_vaccinated_per_hundred'].last()
latest.to_excel('excel_files/vax_summary.xlsx')
```
